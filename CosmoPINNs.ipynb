```python
import json
import html
from IPython.display import display, HTML

with open("config.json", "r", encoding="utf-8") as f:
    data = json.load(f)

data["eps_global"] = -4.0
data["lambda1"] = 60.0
data["lambda2"] = 0.8

data["enable_phase2"] = False
data["enable_phase3"] = True

data["use_clip_gn_phase2"] = False
data["use_clip_gn_phase3"] = True

data["run_phase2_only"] = False
data["run_phase3_only"] = True

with open("config.json", "w", encoding="utf-8") as f:
    json.dump(data, f, indent=2, ensure_ascii=False)

text = json.dumps(data, indent=2, ensure_ascii=False)

highlight_keys = ["eps_global",
                  "lambda1", "lambda2",
                  "enable_phase2", "enable_phase3",
                  "use_clip_gn_phase2", "use_clip_gn_phase3",
                  "run_phase2_only", "run_phase3_only"]

lines = []
for line in text.splitlines():
    escaped = html.escape(line)
    if any(f'"{k}":' in line for k in highlight_keys):
        lines.append(f'<span style="color:red; font-weight:bold;">{escaped}</span>')
    else:
        lines.append(escaped)

display(HTML("<pre>" + "<br>".join(lines) + "</pre>"))
```
